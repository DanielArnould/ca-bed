```python
from typing import TypedDict


class VictimInformation(TypedDict):
    name: str
    introduction: str
    cause_of_death: str
    murder_weapon: str


class SuspectInformation(TypedDict):
    name: str
    introduction: str
    relationship: str
    reason_at_scene: str
    suspicion: str
    motive: str
    opportunity: str
    evidence: str
    testimony: str
    is_murderer: bool


class DetectiveCasesInstance(TypedDict):
    num: int
    time: str
    location: str
    victim: VictimInformation
    suspects: list[SuspectInformation]
```

```python
import json
import random


with open("ARBenchTest.json", "r") as f:
    d = json.load(f)

cases = []
for i, case in enumerate(d):
    time = case["time"]
    location = case["location"]
    victim = case["victim"]
    suspects = []

    for suspect in case["suspects"]:
        if "is_murderer" not in suspect:
            continue
        name = suspect["name"]
        introduction = suspect["introduction"]
        relationship = suspect["relationship"]
        reason_at_scene = suspect["reason_at_scene"][0]
        suspicion = suspect["suspicion"][0]
        motive = suspect["motive"][0]
        opportunity = suspect["opportunity"][0]
        evidence = suspect["evidence"]
        testimony = suspect["testimony"]
        is_murderer = suspect["is_murderer"]

        suspects.append(
            {
                "name": name,
                "introduction": introduction,
                "relationship": relationship,
                "reason_at_scene": reason_at_scene,
                "suspicion": suspicion,
                "motive": motive,
                "opportunity": opportunity,
                "evidence": evidence,
                "testimony": testimony,
                "is_murderer": is_murderer,
            }
        )

    random.shuffle(suspects)
    cases.append(
        {
            "num": i,
            "time": time,
            "location": location,
            "victim": victim,
            "suspects": suspects,
        }
    )
```

```python
with open("DetectiveCases.json", "w") as f:
    json.dump(cases, f)
```
